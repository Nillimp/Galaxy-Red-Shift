```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import ElasticNet
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
file_path = 'SDSS_DR18_GALAXY.csv'
data = pd.read_csv(file_path)
data.head()
# data.tail()
```

```python
X = data[['u', 'g', 'r', 'i', 'z']]
y = data['redshift']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
model_enet = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)

scoring_rmse = make_scorer(mean_squared_error, squared=False)
scoring_r2 = 'r2'
```

```python
rmse_cv = cross_val_score(model_enet, X_scaled, y, cv=kf, scoring=scoring_rmse)
r2_cv = cross_val_score(model_enet, X_scaled, y, cv=kf, scoring=scoring_r2)

mean_rmse_cv = rmse_cv.mean()
mean_r2_cv = r2_cv.mean()

print("Mean RMSE (Elastic Net, Cross-Validation):", mean_rmse_cv)
print("Mean R-squared (Elastic Net, Cross-Validation):", mean_r2_cv)
```
