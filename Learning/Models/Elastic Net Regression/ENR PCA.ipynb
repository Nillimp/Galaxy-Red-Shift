```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
file_path = 'SDSS_DR18_GALAXY.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
X = data[['u', 'g', 'r', 'i', 'z']]
y = data['redshift']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)  
X_pca = pca.fit_transform(X_scaled)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)
```

```python
model_enet = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)

model_enet.fit(X_train_pca, y_train)
```

```python
y_pred_pca = model_enet.predict(X_test_pca)
```

```python
rmse_enet_pca = mean_squared_error(y_test, y_pred_pca, squared=False)  
r2_enet_pca = r2_score(y_test, y_pred_pca)

print("RMSE (Elastic Net with PCA):", rmse_enet_pca)
print("R-squared (Elastic Net with PCA):", r2_enet_pca)
```
