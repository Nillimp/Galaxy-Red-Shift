```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('SDSS_DR18_filtered.csv')
data.head()
```

```python
bands = ['u', 'g', 'r', 'i', 'z']
```

```python
corr_matrix = data[bands].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Photometric Bands - Multicollinearity Check')
plt.show()
```
