```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'SDSS_DR18_GALAXY.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
print("Column names in the dataset:")
print(data.columns.tolist())
```

```python
photometric_bands = data[['u', 'g', 'r', 'i', 'z']]
data
```

```python
plt.figure(figsize=(10, 6))
photometric_bands.boxplot()
plt.title('Boxplot of Photometric Bands (u, g, r, i, z)')
plt.ylabel('Magnitude')
plt.show()
```

```python
# Display the first few rows of the photometric bands
print(photometric_bands.head())

# Check for unique values in each band to understand their distribution
print("Unique values in photometric bands:")
for band in ['u', 'g', 'r', 'i', 'z']:
    print(f"{band} unique values: {photometric_bands[band].unique()}")
```
